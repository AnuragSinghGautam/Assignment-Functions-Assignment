# **Q1.  What is the difference between a function and a method in Python?**
Ans. In Python, functions are defined using the def keyword and can be called independently, while methods are associated with an object and are called on that object within a class, with methods having access to the object.

**Function Example:**


```
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))  # Output: Hello, Alice!

```

**Method Example:**


```
class Greeter:
    def __init__(self, name):
        self.name = name
    
    def greet(self):
        return f"Hello, {self.name}!"

g = Greeter("Alice")
print(g.greet())  # Output: Hello, Alice!

```

# **5. What are iterators in Python and how do they differ from iterables?**
Ans.In Python, iterators and iterables are related concepts but have distinct roles.

**Iterables:**

Iterables are Python objects that return members one at a time, like lists, tuples, strings, and dictionaries, and implement the __iter__() method.

**Example:**


```
# A list is an iterable
numbers = [1, 2, 3, 4]
for number in numbers:
    print(number)

```
**Iterators:**

An iterator is a data stream object that iterates over iterables, producing items one at a time. It has two methods: __iter__() returns the iterator object, and __next__() returns the next item.

**Example:**


```
# Create an iterator from a list
numbers = [1, 2, 3, 4]
iterator = iter(numbers)  # Create an iterator

# Use the iterator
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2

```

**Key Differences:**

An iterable is an object that can be looped over using functions like map() or filter(). It provides an iterator, which consumes elements one at a time and maintains its current position. Iterators are the objects that perform the iteration, while iterables are objects that can access elements in a sequence.

# **6. Explain the concept of generators in Python and how they are defined.**
Ans.Generators are a Python iterable type that allows for iteration using a function that yields values one at a time, using the yield statement. This allows the function to return a value and pause execution, maintaining its state for the next iteration. To define a generator, use a normal function with one or more yield statements.

**Example:**


```
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

# Using the generator
for number in count_up_to(5):
    print(number)  # Output: 1, 2, 3, 4, 5

```



# **7. What are the advantages of using generators over regular functions?**
Ans. Generators offer memory efficiency by yielding one item at a time, reducing the need for storage of all values simultaneously. They also compute values on-the-fly, allowing them to start yielding values before the entire set is ready. This simplifies code for iteration, reducing boilerplate and improving performance, especially with large datasets.

**Example:**



```
def generate_squares(n):
    for i in range(n):
        yield i * i

for square in generate_squares(5):
    print(square)  # Output: 0, 1, 4, 9, 16

```

# **10. What is the difference between `map()`, `reduce()`, and `filter()`functions in Python?**
Ans.The functions map(), filter(), and reduce() are used to apply a function to each item in an iterable, filter() based on a function that returns True or False, and reduce() cumulatively applies a function to the items of an iterable, reducing it to a single value, which requires importation from functools.

map():


```
# Example: Doubling each number in a list
numbers = [1, 2, 3, 4]
doubled = map(lambda x: x * 2, numbers)
print(list(doubled))  # Output: [2, 4, 6, 8]
```

filter():



```
# Example: Filtering even numbers from a list
numbers = [1, 2, 3, 4]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  # Output: [2, 4]
```

reduce():



```
from functools import reduce

# Example: Summing all numbers in a list
numbers = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, numbers)
print(total)  # Output: 10

```
map(): Transforms each item in an iterable.

filter(): Filters items in an iterable based on a condition.

reduce(): Reduces an iterable to a single cumulative value.

# **5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**


````python
def func(num):
  a,b=0,1
  for i in range(num):
    yield a
    a,b=b,a+b
for i in func(10):
  print(i)
````

# **7. Implement a generator function that reads a file line by line and yields each line as a string**

````python
def func(string):
    i = 0
    while i < len(string):
        if string[i] == '\n':
            yield string[:i]
            string = string[i+1:]
            i = 0
        else:
            i += 1
    yield string

for line in func("hello there human.\n my name is abcd"):
    print(line)

````

# **9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

````python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temperatures = [0, 20, 37, 100]
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(fahrenheit_temperatures)
````

# **10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**

````python
def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_string))

input_string = "Hello, World!"
no_vowels_string = remove_vowels(input_string)

print(no_vowels_string)
````
